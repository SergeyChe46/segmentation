# Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Read image

```python
sample_image = cv2.imread('image.jpg')
image = cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

# Preprocessing image

```python
image = cv2.resize(image, (256, 256))
two_dim_image = image.reshape((-1, 3))
two_dim_image = np.float32(two_dim_image)
```

## Define parameters of K

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 2
attempts = 10
```

### Apply K-means to image

```python
ret, label, center = cv2.kmeans(two_dim_image, K, None, criteria, attempts, cv2.KMEANS_PP_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
result_image = res.reshape((image.shape))

plt.axis('off')
plt.imshow(result_image)
```

```python
ret, label, center = cv2.kmeans(two_dim_image, 3, None, criteria, attempts, cv2.KMEANS_PP_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
result_image = res.reshape((image.shape))

plt.axis('off')
plt.imshow(result_image)
```
